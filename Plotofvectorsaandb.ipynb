```python
import matplotlib.pyplot as plt
import numpy as np

# Magnitudes of the vectors
magnitude_a = 3
magnitude_b = 4

# Angles in degrees
angle_a = 30
angle_b = 60

# Convert angles to radians
angle_a_rad = np.deg2rad(angle_a)
angle_b_rad = np.deg2rad(angle_b)

# Compute the components of the vectors
a_x = magnitude_a * np.cos(angle_a_rad)
a_y = magnitude_a * np.sin(angle_a_rad)

b_x = magnitude_b * np.cos(angle_b_rad)
b_y = magnitude_b * np.sin(angle_b_rad)

# Plot the vectors
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, a_x, a_y, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
plt.quiver(0, 0, b_x, b_y, angles='xy', scale_units='xy', scale=1, color='red', label='Vector b')

# Set plot limits
plt.xlim(-1, 5)
plt.ylim(-1, 5)

# Add labels, legend, and grid
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.title('Plot of Vectors a and b')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
r = 5  # Magnitude of vector r
theta = 45  # Angle of the original vector (in degrees)
rotation_angle = 30  # Rotation angle of the plane (in degrees)

# Convert angles to radians
theta_rad = np.deg2rad(theta)
rotation_rad = np.deg2rad(rotation_angle)

# Original vector components
x_original = r * np.cos(theta_rad)
y_original = r * np.sin(theta_rad)

# Transformed vector components using rotation formula
x_transformed = x_original * np.cos(rotation_rad) - y_original * np.sin(rotation_rad)
y_transformed = x_original * np.sin(rotation_rad) + y_original * np.cos(rotation_rad)

# Plotting
plt.figure(figsize=(8, 8))
plt.axhline(0, color='gray', linewidth=0.5)  # X-axis
plt.axvline(0, color='gray', linewidth=0.5)  # Y-axis

# Plot original vector
plt.quiver(0, 0, x_original, y_original, angles='xy', scale_units='xy', scale=1, color='blue', label='Original Vector r')
# Plot transformed vector
plt.quiver(0, 0, x_transformed, y_transformed, angles='xy', scale_units='xy', scale=1, color='green', label='Transformed Vector r\'')

# Add labels for the vectors
plt.text(x_original * 0.5, y_original * 0.5, "r", color="blue", fontsize=12, ha="center", va="center")
plt.text(x_transformed * 0.5, y_transformed * 0.5, "r'", color="green", fontsize=12, ha="center", va="center")

# Set limits and grid
plt.xlim(-r - 1, r + 1)
plt.ylim(-r - 1, r + 1)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')  # Keep aspect ratio equal

# Add title and legend
plt.title('Original and Transformed Vectors')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

```python
pip install streamlit

import streamlit as st
import openai
import uuid
from graphviz import Source
from pptx import Presentation
from pptx.util import Inches
import tempfile
import os
import re

# Set OpenAI API key
openai.api_key = st.secrets.get("OPENAI_API_KEY") or os.getenv("OPENAI_API_KEY")

st.set_page_config(page_title="Diagram Idea Generator from Blog", layout="wide")
st.title("AI-Powered Blog-to-Diagram Idea Generator")

st.markdown("""
This app helps you extract **diagram ideas** from long-form blog content.
Instead of trying to figure out manually what visuals might support your writing, this tool uses GPT-4 to generate **visualizable ideas** based on your blog's logic, steps, or message.

🧠 **Use this to:**
- Get visual inspiration for your content
- Enrich blog posts with relevant diagrams
- Support slide decks or product explainers
- Jumpstart client deliverables or interview presentations

You don’t need to create diagrams in this app. Instead, **use the extracted ideas as your visual planning tool**.
""")

# Initialize session state
if "diagram_input" not in st.session_state:
    st.session_state.diagram_input = ""
if "ideas_list" not in st.session_state:
    st.session_state.ideas_list = []
if "selected_title" not in st.session_state:
    st.session_state.selected_title = ""

# Examples dropdown to guide users
example_options = {
    "Data Science Workflow": "A typical machine learning workflow starts with data collection, followed by preprocessing, feature engineering, model training, evaluation, and deployment.",
    "Product Launch Strategy": "Launching a new product involves ideation, user research, prototype development, internal testing, public beta, feedback collection, final rollout, and marketing.",
    "Content Creation Process": "Writing a long-form blog post includes topic selection, outline creation, research, drafting, editing, formatting, SEO optimization, and publishing."
}

with st.expander("\U0001F4A1 See Example Blog Inputs & Tones"):
    example_choice = st.selectbox("Choose an example to autofill:", list(example_options.keys()))
    if st.button("Use Example Content"):
        st.session_state.example_input = example_options[example_choice]
    st.markdown("""
    **Suggested Tones for Blog Output:**
    - Friendly
    - Technical/Precise
    - Inspirational
    - Conversational
    - Thought-leadership
    """)
    selected_tone = st.selectbox("Choose a tone for your blog post:", ["Friendly", "Technical/Precise", "Inspirational", "Conversational", "Thought-leadership"])

# Visual Theme and Diagram Type Selections
visual_theme = st.selectbox("Choose Visual Theme", ["Default", "Dark", "Pastel", "Monochrome"])
diagram_type = st.selectbox("Choose Diagram Type", ["Linear", "Hierarchical", "Cycle", "Cluster"])

# Input prompt
user_input = st.text_area("Paste your blog post or article content:", value=st.session_state.get("example_input", ""), placeholder="Paste your full blog content here to generate visual ideas from it")

def extract_multiple_diagram_ideas(text):
    prompt = f"""
    Read the following blog post and extract 5 to 10 processes or conceptual ideas that could be visualized as diagrams. For each idea, write a short summary followed by a simple arrow-based structure.

    Format:
    1. Title: [Name of Idea]
       Diagram: Step1 -> Step2 -> Step3

    Blog Content:
    {text}
    """
    response = openai.chat.completions.create(
        model="gpt-4-1106-preview",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.5,
        max_tokens=800
    )
    return response.choices[0].message.content.strip()

st.subheader("✨ Get Visual Thinking Prompts from Your Blog")

if st.button("Extract Diagram Ideas from Blog") and user_input:
    with st.spinner("Extracting visual ideas..."):
        extracted_ideas = extract_multiple_diagram_ideas(user_input)
        st.markdown("#### Full GPT Response (for reference)")
        st.code(extracted_ideas, language="text")
        st.markdown("### Detected Diagram Concepts:")
        st.session_state.ideas_list = []
        blocks = re.split(r'\n\d+\.', extracted_ideas)
        for block in blocks:
            if 'Diagram:' in block:
                title_line = next((line for line in block.splitlines() if 'Title:' in line), "Untitled")
                diagram_line = next((line for line in block.splitlines() if 'Diagram:' in line), None)
                if diagram_line:
                    st.session_state.ideas_list.append((title_line.strip(), diagram_line.replace("Diagram:", "").strip()))

if st.session_state.ideas_list:
    idea_options = [title for title, _ in st.session_state.ideas_list]
    selected_idea = st.selectbox("Choose a diagram idea to visualize:", idea_options)
    if st.button("Use This Idea for Diagram"):
        st.session_state.selected_title = selected_idea
        st.session_state.diagram_input = dict(st.session_state.ideas_list).get(selected_idea, "")

st.markdown("---")
st.subheader("Optional: Visualize a Selected Diagram")

if st.button("Generate Diagram") and st.session_state.diagram_input:
    with st.spinner("Generating diagram..."):
        diagram_text = st.session_state.diagram_input
        st.markdown(f"**Using Idea: {st.session_state.selected_title}**")
        st.code(diagram_text or "[No structure extracted]", language="text")

        def sanitize_diagram_text(diagram_text):
            parts = re.split(r'->', diagram_text)
            cleaned_parts = [f'"{p.strip()}"' for p in parts if p.strip()]
            return ' -> '.join(cleaned_parts)

        cleaned_diagram = sanitize_diagram_text(diagram_text)

        style_map = {
            "Default": "",
            "Dark": "bgcolor=black; fontcolor=white; node [color=white, fontcolor=white];",
            "Pastel": "node [style=filled, fillcolor=lightblue];",
            "Monochrome": "node [style=filled, fillcolor=gray];"
        }
        layout_map = {
            "Linear": "LR",
            "Hierarchical": "TB",
            "Cycle": "LR",
            "Cluster": "LR"
        }
        rankdir = layout_map.get(diagram_type, "LR")
        style = style_map.get(visual_theme, "")
        dot_code = f'digraph G {{ rankdir={rankdir}; {style} {cleaned_diagram} }}'

        src = Source(dot_code)
        filename = f"diagram_{uuid.uuid4().hex}"
        png_path = os.path.join(tempfile.gettempdir(), f"{filename}.png")
        src.render(filename=png_path[:-4], format="png", cleanup=True)
        st.image(png_path, caption="Blog-Ready Smart Visual")

        with open(png_path, "rb") as img_file:
            st.download_button(
                label="Download Blog Visual as PNG",
                data=img_file,
                file_name="smart_diagram.png",
                mime="image/png"
            )

        if st.button("Export to PowerPoint"):
            prs = Presentation()
            slide_layout = prs.slide_layouts[5]
            slide = prs.slides.add_slide(slide_layout)
            title = slide.shapes.title
            title.text = "Blog Smart Visual"
            slide.shapes.add_picture(png_path, Inches(1), Inches(2), height=Inches(4))
            pptx_file = tempfile.NamedTemporaryFile(delete=False, suffix=".pptx")
            prs.save(pptx_file.name)
            with open(pptx_file.name, "rb") as f:
                st.download_button("Download PowerPoint", f, "diagram_slide.pptx")
```
