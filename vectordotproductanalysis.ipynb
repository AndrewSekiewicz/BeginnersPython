```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)

# Add a new column
df['Salary'] = [70000, 80000, 65000, 90000]
print("\nDataFrame with Salary column added:")
print(df)

# Filter rows where Age is greater than 25
filtered_df = df[df['Age'] > 25]
print("\nFiltered DataFrame (Age > 25):")
print(filtered_df)

# Group by 'City' and calculate the mean age
grouped_df = df.groupby('City')['Age'].mean()
print("\nMean Age by City:")
print(grouped_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def visualize_2d_dot_product():
    # Simple 2D vectors
    a = np.array([4, 3])
    b = np.array([2, 1])

    # Calculate dot product and related values
    dot_product = np.dot(a, b)
    magnitude_a = np.linalg.norm(a)
    magnitude_b = np.linalg.norm(b)
    cos_theta = dot_product / (magnitude_a * magnitude_b)
    angle_rad = np.arccos(np.clip(cos_theta, -1, 1))
    angle_deg = np.degrees(angle_rad)

    # Create figure with subplots
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

    # Plot 1: Basic vectors
    ax1.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1,
               color='red', width=0.008, label=f'a = ({a[0]}, {a[1]})')
    ax1.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1,
               color='blue', width=0.008, label=f'b = ({b[0]}, {b[1]})')

    # Add grid and formatting
    ax1.grid(True, alpha=0.3)
    ax1.set_xlim(-1, 5)
    ax1.set_ylim(-1, 4)
    ax1.set_xlabel('X')
    ax1.set_ylabel('Y')
    ax1.set_title('2D Vectors')
    ax1.legend()
    ax1.set_aspect('equal')

    # Add origin point
    ax1.plot(0, 0, 'ko', markersize=8)

    # Plot 2: Angle between vectors
    ax2.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1,
               color='red', width=0.008, label=f'a = ({a[0]}, {a[1]})')
    ax2.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1,
               color='blue', width=0.008, label=f'b = ({b[0]}, {b[1]})')

    # Draw angle arc
    angle_a = np.arctan2(a[1], a[0])
    angle_b = np.arctan2(b[1], b[0])

    # Create arc between vectors
    angles = np.linspace(min(angle_a, angle_b), max(angle_a, angle_b), 50)
    arc_radius = 0.8
    arc_x = arc_radius * np.cos(angles)
    arc_y = arc_radius * np.sin(angles)
    ax2.plot(arc_x, arc_y, 'g-', linewidth=2, label=f' = {angle_deg:.1f}')

    ax2.grid(True, alpha=0.3)
    ax2.set_xlim(-1, 5)
    ax2.set_ylim(-1, 4)
    ax2.set_xlabel('X')
    ax2.set_ylabel('Y')
    ax2.set_title(f'Angle Between Vectors: {angle_deg:.1f}')
    ax2.legend()
    ax2.set_aspect('equal')
    ax2.plot(0, 0, 'ko', markersize=8)

    # Plot 3: Vector projection
    # Project b onto a
    proj_b_on_a = (dot_product / (magnitude_a ** 2)) * a

    ax3.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1,
               color='red', width=0.008, label=f'a = ({a[0]}, {a[1]})')
    ax3.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1,
               color='blue', width=0.008, label=f'b = ({b[0]}, {b[1]})')
    ax3.quiver(0, 0, proj_b_on_a[0], proj_b_on_a[1], angles='xy', scale_units='xy', scale=1,
               color='green', width=0.006, label=f'proj_a(b)')

    # Draw projection line (perpendicular from b to projection)
    ax3.plot([b[0], proj_b_on_a[0]], [b[1], proj_b_on_a[1]],
             'g--', alpha=0.7, linewidth=1, label='perpendicular')

    ax3.grid(True, alpha=0.3)
    ax3.set_xlim(-1, 5)
    ax3.set_ylim(-1, 4)
    ax3.set_xlabel('X')
    ax3.set_ylabel('Y')
    ax3.set_title('Projection of b onto a')
    ax3.legend()
    ax3.set_aspect('equal')
    ax3.plot(0, 0, 'ko', markersize=8)

    # Plot 4: Calculations and results
    ax4.axis('off')

    calculation_text = f"""
DOT PRODUCT CALCULATION

Vectors:
a = ({a[0]}, {a[1]})
b = ({b[0]}, {b[1]})

Step-by-step calculation:
a  b = ab + ab
a  b = {a[0]}{b[0]} + {a[1]}{b[1]}
a  b = {a[0]*b[0]} + {a[1]*b[1]} = {dot_product}

Magnitudes:
|a| = ({a[0]} + {a[1]}) = {a[0]**2 + a[1]**2} = {magnitude_a:.2f}
|b| = ({b[0]} + {b[1]}) = {b[0]**2 + b[1]**2} = {magnitude_b:.2f}

Angle calculation:
cos() = (ab)/(|a||b|)
cos() = {dot_product}/({magnitude_a:.2f}{magnitude_b:.2f})
cos() = {cos_theta:.3f}
 = {angle_deg:.1f}

Projection:
proj_a(b) = ((ab)/|a|)  a
proj_a(b) = ({dot_product}/{magnitude_a**2:.1f})  ({a[0]}, {a[1]})
proj_a(b) = ({proj_b_on_a[0]:.2f}, {proj_b_on_a[1]:.2f})

INTERPRETATION:
 Dot product = {dot_product} (positive)
 Vectors point in similar directions
 Angle = {angle_deg:.1f} (acute angle < 90)
 This indicates the vectors are "aligned"
    """

    ax4.text(0.05, 0.95, calculation_text, transform=ax4.transAxes,
             fontsize=9, verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

    plt.tight_layout()
    plt.show()

    # Print summary
    print("SIMPLE 2D DOT PRODUCT EXAMPLE")
    print("=" * 40)
    print(f"Vector a = {a}")
    print(f"Vector b = {b}")
    print(f"Dot product ab = {dot_product}")
    print(f"Angle between vectors = {angle_deg:.1f}")
    print(f"Interpretation: {'Vectors point in similar directions' if dot_product > 0 else 'Vectors point in opposite directions'}")

# Run the visualization
if __name__ == "__main__":
    visualize_2d_dot_product()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patches as patches

def visualize_dot_product():
    # Define vectors from our example
    u = np.array([3, -2, 1])
    v = np.array([1, 4, -2])

    # Calculate dot product and related values
    dot_product = np.dot(u, v)
    magnitude_u = np.linalg.norm(u)
    magnitude_v = np.linalg.norm(v)
    cos_theta = dot_product / (magnitude_u * magnitude_v)
    angle_rad = np.arccos(np.clip(cos_theta, -1, 1))
    angle_deg = np.degrees(angle_rad)

    # Create figure with subplots
    fig = plt.figure(figsize=(15, 10))

    # 3D plot
    ax1 = fig.add_subplot(221, projection='3d')

    # Plot vectors
    ax1.quiver(0, 0, 0, u[0], u[1], u[2], color='red', arrow_length_ratio=0.1,
               linewidth=3, label=f'u = {u}')
    ax1.quiver(0, 0, 0, v[0], v[1], v[2], color='blue', arrow_length_ratio=0.1,
               linewidth=3, label=f'v = {v}')

    # Add origin point
    ax1.scatter([0], [0], [0], color='black', s=50)

    # Set labels and title
    ax1.set_xlabel('X')
    ax1.set_ylabel('Y')
    ax1.set_zlabel('Z')
    ax1.set_title('3D Vector Visualization')
    ax1.legend()

    # Set equal aspect ratio
    max_range = max(max(abs(u)), max(abs(v))) * 1.2
    ax1.set_xlim([-max_range, max_range])
    ax1.set_ylim([-max_range, max_range])
    ax1.set_zlim([-max_range, max_range])

    # 2D projection (XY plane)
    ax2 = fig.add_subplot(222)
    ax2.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1,
               color='red', width=0.005, label=f'u (XY projection)')
    ax2.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1,
               color='blue', width=0.005, label=f'v (XY projection)')

    # Draw angle arc
    if abs(angle_deg) > 1:  # Only draw if angle is significant
        angle_arc = patches.Arc((0, 0), 1.5, 1.5, angle=0,
                               theta1=0, theta2=np.degrees(np.arctan2(u[1], u[0])),
                               color='green', linewidth=2)
        ax2.add_patch(angle_arc)
        angle_arc2 = patches.Arc((0, 0), 1.5, 1.5, angle=0,
                                theta1=0, theta2=np.degrees(np.arctan2(v[1], v[0])),
                                color='green', linewidth=2, linestyle='--')
        ax2.add_patch(angle_arc2)

    ax2.set_xlim([-max_range, max_range])
    ax2.set_ylim([-max_range, max_range])
    ax2.set_xlabel('X')
    ax2.set_ylabel('Y')
    ax2.set_title('XY Plane Projection')
    ax2.grid(True, alpha=0.3)
    ax2.set_aspect('equal')
    ax2.legend()

    # Vector projection visualization (3D)
    ax3 = fig.add_subplot(223, projection='3d')

    # Project v onto u
    proj_v_on_u = (dot_product / (magnitude_u ** 2)) * u

    # Draw vectors for projection
    ax3.quiver(0, 0, 0, u[0], u[1], u[2], color='red', arrow_length_ratio=0.1,
               linewidth=3, label=f'u = {u}')
    ax3.quiver(0, 0, 0, v[0], v[1], v[2], color='blue', arrow_length_ratio=0.1,
               linewidth=3, label=f'v = {v}')
    ax3.quiver(0, 0, 0, proj_v_on_u[0], proj_v_on_u[1], proj_v_on_u[2],
               color='green', arrow_length_ratio=0.1, linewidth=2,
               label=f'proj_u(v)')

    # Draw projection line
    ax3.plot([v[0], proj_v_on_u[0]], [v[1], proj_v_on_u[1]], [v[2], proj_v_on_u[2]],
             'g--', alpha=0.7, linewidth=1)

    # Set equal aspect ratio for projection plot
    ax3.set_xlim([-max_range, max_range])
    ax3.set_ylim([-max_range, max_range])
    ax3.set_zlim([-max_range, max_range])

    ax3.set_xlabel('X')
    ax3.set_ylabel('Y')
    ax3.set_zlabel('Z')
    ax3.set_title('Vector Projection')
    ax3.legend()

    # Text summary
    ax4 = fig.add_subplot(224)
    ax4.axis('off')

    summary_text = f"""
    VECTOR DOT PRODUCT ANALYSIS

    Vectors:
    u = {u}
    v = {v}

    Calculations:
    u  v = {u[0]}{v[0]} + ({u[1]}){v[1]} + {u[2]}({v[2]})
    u  v = {u[0]*v[0]} + {u[1]*v[1]} + {u[2]*v[2]} = {dot_product}

    Magnitudes:
    |u| = ({u[0]} + ({u[1]}) + {u[2]}) = {magnitude_u:.3f}
    |v| = ({v[0]} + {v[1]} + ({v[2]})) = {magnitude_v:.3f}

    Angle between vectors:
    cos() = (uv)/(|u||v|) = {dot_product}/({magnitude_u:.3f}{magnitude_v:.3f})
    cos() = {cos_theta:.3f}
     = {angle_deg:.1f}

    Interpretation:
     Dot product is {dot_product} (negative)
     Vectors point in generally opposite directions
     Angle > 90 indicates obtuse angle

    Projection of v onto u:
    proj_u(v) = ((uv)/|u|)  u = {proj_v_on_u}
    """

    ax4.text(0.05, 0.95, summary_text, transform=ax4.transAxes, fontsize=10,
             verticalalignment='top', fontfamily='monospace',
             bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()
    plt.show()

    # Additional geometric interpretation
    print("GEOMETRIC INTERPRETATION:")
    print(f" The dot product {dot_product} tells us the vectors are pointing")
    print("  in generally opposite directions (negative value)")
    print(f" The angle of {angle_deg:.1f} is obtuse (> 90)")
    print(f" If we project vector v onto vector u, we get: {proj_v_on_u}")
    print(f" The magnitude of this projection is: {np.linalg.norm(proj_v_on_u):.3f}")

    return dot_product, angle_deg, proj_v_on_u

# Run the visualization
if __name__ == "__main__":
    result = visualize_dot_product()
```
